```python
import pandas as pd
# import numpy as np
# import datetime as dt
```

```python
#imports the data frame- UPS data from the previous month
#pulls off the first 82 columns
#keeps only the columns needed for the analysis

df = pd.read_excel('UPS R1 February 2021.xlsx', header=None)
df2 = df.iloc[:, [2,4,5,10,11,13,16,22,45,52,75,76,78,79,80]]

df2.head()
```

```python
#names headers for the columns

df2.columns = ["account", "invoice_date", "invoice_no", "invoice_total", "ship_date", "tracking_no", "ref_1", "ref_2", "charge_type", "final_charge","receiving_party", "receiving_address", "city", "state", "zip" ]
```

```python
#inserts a column for the sales order number but with 0 in each row 

df2.insert(6, "sales_order_no", str(0))
```

```python
df2.head()
```

```python
df2.insert(7, "card_code", '')
```

```python
#creates a new column called "right" which pulls off the sales order no from ref field. However need to put this in the
#sales order column

right = df2['ref_1'].str[-6:]
```

```python
#itterates through each row and makes the column "sales order no" equal to the right 6 from the ref column so we have 
#SO number in separate column. 

for i, row in df.iterrows():
    df2['sales_order_no'] = right
    
display(df2)
```

```python
for i, row in df2.iterrows():
    df2['sales_order_no'] = df2['sales_order_no'].apply(pd.to_numeric, errors = 'coerce')
  
```

```python
#drops the second row which is a numbered row, not needed. Need to see if it's working. 

df2.drop(df2.index[0])
```

```python
#converts the column "invoice date" to a datetime object

df2['invoice_date'] = pd.to_datetime(df2['invoice_date'], errors = 'coerce')
```

```python
#converts invoice date to a date object

df2['invoice_date'] = df2['invoice_date'].dt.date
```

```python
#converts the column "ship date" to a datetime object

df2['ship_date'] = pd.to_datetime(df2['ship_date'], errors = 'coerce')
```

```python
#converts ship date to a date object

df2['ship_date'] = df2['ship_date'].dt.date
df2.dtypes
```

```python
#saves new dataframe excel doc. 

df2.to_excel('UPS R1 February condensed.xlsx', index = False)
```

MERGE DATA FRAMES (UPS RAW DATA WITH ORDER AND LINE ITEMS)

```python
#pull in ordr sheet

df_ordr = pd.read_excel('ORDR for UPS analysis.xlsx', sheet_name = 'ORDR_only')

df_ordr.head()
```

```python
#pull in ordr line items sheet 

df_lineitems = pd.read_excel('ORDR for UPS analysis.xlsx', sheet_name = 'ORDR_line_items')

df_lineitems.head()
```

```python
#merge ORDR and UPS data dataframes

df_merge = pd.merge(df2, df_ordr, on = ['sales_order_no'], how = 'left')
```

```python
df_merge.head()
```

```python
#push the new DF to excel file

df_merge.to_excel('df_merge.xlsx', index = False)
```

NOW START THE DATA ANALYSIS

```python
#find the mean & shape of df_merge, also count unique tracking numbers

print(df_merge.final_charge.mean()) 
print(df_merge.shape)
print(df_merge.tracking_no.value_counts()) 
```

```python
#drop the rows where final charge = 0

df_merge = df_merge[df_merge['final_charge'] != 0]
```

```python
#print the shape after dropping null charge rows, find the new mean without the null rows. 
#this is finding the mean of all charges, but we need to sum on tracking numbers and find the mean of the shipping charge
#aka take sum final charge divide by # of unique tracking. 

print(df_merge.shape)
print(df_merge.final_charge.mean())
```

```python
#sum final charge by invoice number and date

invoice_by_date = df_merge.groupby(['invoice_date', 'invoice_no'])[['final_charge']].sum()
print(invoice_by_date)
```

```python
#creates a list called "packages" that is the unique tracking numbers
#counts number of items in that list
#show sum of final charges
#returns the average cost per trakcing number based on final_charge

packages = df_merge.tracking_no.value_counts()
total_charges = df_merge.final_charge.sum() 

print(packages.count())
print(total_charges)
print(total_charges/packages.count())
```
